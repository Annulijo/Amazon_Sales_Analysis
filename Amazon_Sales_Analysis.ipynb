VISUALIZATION


```python
sns.countplot(x=df['Item Type'])
plt.xlabel('Item Type')
plt.ylabel('Count')
plt.title('Count of Items')
plt.xticks(rotation=90)
plt.show()
```

Total Revenue by Sales Channel

```python
sales_cha = df.groupby('Sales Channel')['Total Revenue'].sum()
plt.figure(figsize=(6, 6))
# sns.barplot(x = sales_cha.index, y = sales_cha.values, palette="magma")
# plt.title('Total Revenue by Sales Channel')
# plt.xlabel('Sales Channel')
# plt.ylabel('Total Revenue')
sales_cha.plot.pie(autopct='%1.1f%%', startangle=90, figsize=(10, 6))
plt.title('Total Revenue by Sales Channel')
plt.tight_layout()
```

Total Revenue by Item Type

```python
item = df.groupby('Item Type')['Total Revenue'].sum()
plt.figure(figsize=(8, 6))
sns.barplot(x = item.index, y = item.values, palette="magma")
plt.title('Total Revenue by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
```

Total Revenue by Region

```python
reg = df.groupby('Region')['Total Revenue'].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x = reg.index, y = reg.values, palette="magma")
plt.title('Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)
```

```python
#Sub-Saharan Africa has the highest total revenue
```

Year Wise Sales


```python
#Converting the order_date and ship_date to datetime Formate
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['sSip Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Adding Year column from the order_date column in the dataset
df['year'] = df['Order Date'].dt.year
```

```python
yearly_sales = df.groupby('year')['Total Revenue'].sum()
plt.figure(figsize=(8, 6))
sns.barplot(x = yearly_sales.index, y = yearly_sales.values, palette="magma")
plt.title('Year Wise Sales')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
```

```python
#Amazon has Maximum Revenue in the year 2012 and Minimum Revenue in the year 2011
```

Month Wise Sales

```python
# Adding Month column from the order_date column in the dataset
df['month'] = df['Order Date'].dt.month
```

```python
monthly_sales = df.groupby('month')['Total Revenue'].sum()
plt.figure(figsize=(8, 6))
sns.barplot(x = monthly_sales.index, y = monthly_sales.values, palette="magma")
plt.title('Month Wise Sales')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
```

```python
#Amazon has Maximum Revenue in the month February and Minimum Revenue in the month August
```

Total Cost by Region

```python
cost_region = df.groupby('Region')['Total Cost'].sum()
plt.figure(figsize=(9, 6))
sns.barplot(x = cost_region.index, y = cost_region.values, palette="magma")
plt.title('Total Cost by Region')
plt.xlabel('Region')
plt.ylabel('Total Cost')
plt.xticks(rotation=90)
```

```python
#The Sub-Saharan Africa has the highest cost followed by Europe and asia
```

Total Cost by Sales Channel

```python
sales_cha = df.groupby('Sales Channel')['Total Cost'].sum()
sales_cha.plot.pie(autopct='%1.1f%%', startangle=90, figsize=(8, 6))
plt.title('Total Cost by Sales Channel')
```

Total Cost by Country

```python
cost_country = df.groupby('Country')['Total Cost'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8, 6))
sns.barplot(x = cost_country.index, y = cost_country.values, palette="magma")
plt.title('Total Cost by Country')
plt.xlabel('Country')
plt.ylabel('Total Cost')
plt.xticks(rotation=90)
```

```python
#Honduras has the highest total cost among all the countries
```

Total units sold of each item tyoe

```python
total_units_sold_item_type = df.groupby('Item Type')['Units Sold'].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x = total_units_sold_item_type.index, y = total_units_sold_item_type.values, palette="Set1")
plt.title('Total Units Sold of each Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45)
plt.tight_layout()
```

Total units sold of each product item monthly

```python
total_units_sold_monthly = df.groupby(['month', 'Item Type'])['Units Sold'].sum().unstack()
plt.figure(figsize=(16, 8))
total_units_sold_monthly.plot(kind='bar', stacked=True)
plt.title('Total Units Sold of each Item Monthly')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
```

```python
#For example in the month of May "Personal Care" products are sold more than any other type of product and snacks are sold very less.
```

Total units sold of each product item yearly

```python
total_units_sold_yearly = df.groupby(['year', 'Item Type'])['Units Sold'].sum().unstack()
plt.figure(figsize=(16, 8))
total_units_sold_monthly.plot(kind='bar', stacked=True)
plt.title('Total Units Sold of each Item Yearly')
plt.xlabel('Year')
plt.ylabel('Total Units Sold')
plt.legend(loc='right')
```

Total Profit by Region

```python
total_profit_by_region = df.groupby('Region')['Total Profit'].sum()
plt.figure(figsize=(10, 5))
sns.barplot(x = total_profit_by_region.index, y = total_profit_by_region.values, palette="magma")
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.title('Total Profit by Region')
plt.xticks(rotation=90)
```

Total Profit by Country

```python
total_profit_by_country = df.groupby('Country')['Total Profit'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x = total_profit_by_country.index, y = total_profit_by_country.values, palette="magma")
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.title('Total Profit by Country')
plt.xticks(rotation=90)
```

Total Profit by Sales Channel

```python
sales_cha1 = df.groupby('Sales Channel')['Total Profit'].sum()
sales_cha1.plot.pie(autopct='%1.1f%%', startangle=90, figsize=(8, 6))
plt.title('Total Profit by Sales Channel')
```

```python
#the maximum profit is generated from Offline sales channel, than online channel.
```

Total profit over time

```python
total_profit_over_time = df.groupby('Order Date')['Total Profit'].sum()
plt.figure(figsize=(13, 6))
total_profit_over_time.plot(kind='line', marker='o', color='k')
plt.xlabel('Order Date')
plt.ylabel('Total Profit')
plt.title('Total Profit over time')
plt.grid(True)
```

Conclusion

Amazon saw its highest revenue in 2012 and lowest in 2011. Djibouti generated the highest total profit. Offline sales channels outperformed online ones, with 56.42% higher profit margins.
